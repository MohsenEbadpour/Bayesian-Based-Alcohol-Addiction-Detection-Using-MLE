```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# read data
all_data = pd.read_csv("blood_test.txt")

# set labale addicted to data
def SetLabel(row):
    if row['drinks'] > 5 :
        return 1
    return 0

all_data['addicted'] = all_data.apply (lambda row: SetLabel(row), axis=1)

# seperate test and train data
train_data = all_data.loc[all_data['selector'] == 1]
train_data = train_data.drop('selector', axis=1)

test_data = all_data.loc[all_data['selector'] == 2]
test_data = test_data.drop('selector', axis=1)

# shffule data 
train_data = train_data.sample(frac=1).reset_index(drop=True)
test_data = test_data.sample(frac=1).reset_index(drop=True)
train_data
```

```python
# not addicted class 
train_data = train_data.sample(frac=1).reset_index(drop=True)
train_data_ = train_data.sample(10)
not_addicted_10_sample = train_data_.loc[train_data['addicted'] == 0].drop('addicted', axis=1)
addicted_10_sample = train_data_.loc[train_data['addicted'] == 1].drop('addicted', axis=1)

train_data = train_data.sample(frac=1).reset_index(drop=True)
train_data_ = train_data.sample(50)
not_addicted_50_sample = train_data_.loc[train_data['addicted'] == 0].drop('addicted', axis=1)
addicted_50_sample = train_data_.loc[train_data['addicted'] == 1].drop('addicted', axis=1)

train_data = train_data.sample(frac=1).reset_index(drop=True)
train_data_ = train_data.sample(100)
not_addicted_100_sample = train_data_.loc[train_data['addicted'] == 0].drop('addicted', axis=1)
addicted_100_sample = train_data_.loc[train_data['addicted'] == 1].drop('addicted', axis=1)
```

```python
# normal distribution parameter with maximum likelihood estimation is the same mean and covariance of dataset

# NOTE
# in previous Homework we immplemented normal distribution calculation, then in this HW i used built-in function in pandas
# to calculate mean and covariance 

print("======> Estimated Parameter for not addicted class <======\n")
print("--> Count of random samples: {0} out of 10".format(not_addicted_10_sample.shape[0]))
print("Means: ",np.array(not_addicted_10_sample.mean()))
print("Covariance: ", np.array(not_addicted_10_sample.cov()))

print("\n--> Count of random samples: {0} out of 50".format(not_addicted_50_sample.shape[0]))
print("Means: ",np.array(not_addicted_50_sample.mean()))
print("Covariance: ", np.array(not_addicted_50_sample.cov()))

print("\n--> Count of random samples: {0} out of 100".format(not_addicted_100_sample.shape[0]))
print("Means: ",np.array(not_addicted_100_sample.mean()))
print("Covariance: ", np.array(not_addicted_100_sample.cov()))
```

```python
print("======> Estimated Parameter for addicted class <======\n")
print("--> Count of random samples: {0} out of 10".format(addicted_10_sample.shape[0]))
print("Means: ",np.array(addicted_10_sample.mean()))
print("Covariance: ", np.array(addicted_10_sample.cov()))

print("\n--> Count of random samples: {0} out of 50".format(addicted_50_sample.shape[0]))
print("Means: ",np.array(addicted_50_sample.mean()))
print("Covariance: ", np.array(addicted_50_sample.cov()))

print("\n--> Count of random samples: {0} out of 100".format(addicted_100_sample.shape[0]))
print("Means: ",np.array(addicted_100_sample.mean()))
print("Covariance: ", np.array(addicted_100_sample.cov()))
```
